```python
import pandas as pd
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
anottation_folder = "Annotations/"
daySequence_folder = "daySequence%i/"
dayTrain_folder = "dayTrain/"
dayClip_folder = "dayClip%i/"
frames_foldeer = "frames/"
```

# dayTrain

```python
dayTrain_df = pd.read_csv(dayTrain_folder+"dayTrain.csv")
dayTrain_df.head()
```

```python
def binaryImage(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    ret, thresh = cv2.threshold(
        gray, 250, 255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
    return thresh
```

```python
image_dataset = []
for index, row in dayTrain_df.iterrows():
    filename = row["Filename"]
    target = row["Annotation tag"]
    img = cv2.imread(dayTrain_folder+target+"/"+filename)
    img = cv2.resize(img, (30, 50))
    img = binaryImage(img)
    img[img == 255] = 1
    img_values = np.c_[img.ravel()].transpose()
    image_info = pd.DataFrame([row.values], columns=["filename", "target", "x1", "y1", "x2", "y2"])
    image_pixels = pd.DataFrame(img_values, columns=["p"+str(i) for i in range(1,1501)])
    image_df = pd.concat([image_info, image_pixels], axis=1)
    print(filename)
    image_dataset.append(image_df)
```

```python
final_df = pd.concat(image_dataset)
final_df.head()
```

```python
final_df.to_csv(dayTrain_folder+"imageDataset.csv", index=False)
```
